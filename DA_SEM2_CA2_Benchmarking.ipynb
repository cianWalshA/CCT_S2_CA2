```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 5000)
pd.set_option('display.max_columns', 5000)
pd.set_option('display.width', 1000)
pd.set_option('display.max_colwidth', None)
import pymongo
```

```python
df = pd.read_csv('/home/hduser/college/assign2/data/databaseBenchmarking.csv')
df['Database'] = df['Test'].str.split(pat='_').str[0]
df['Workload'] = df['Test'].str.split(pat='_').str[1]
```

```python
df. head()
```

```python

fig, (ax1) = plt.subplots(nrows=1, figsize=(15,7))
sns.color_palette("Set2")
sns.set_style("ticks")
ax1 = sns.barplot(data=df[df['Variable']=='RunTime (ms)'], x="Database", y="Result", hue='Workload', palette='Set2', errorbar =None)
ax1.set_title("MySQL vs MongoDB 10,000 records and 10,000 operations - Run Time")
ax1.set(ylabel='Runtime (ms)', xlabel='Database')
ax1.figure.savefig('/mnt/c/Users/cianw/Documents/runtime.png', dpi=1200, facecolor='w')
```

```python

fig, (ax2) = plt.subplots(nrows=1, figsize=(15,7))
sns.color_palette("Set2")
sns.set_style("whitegrid")
ax2 = sns.barplot(data=df[df['Variable']=='Throughput (ops/sec)'], x="Database", y="Result", hue='Workload', palette='Set2', errorbar =None)
ax2.set_title("MySQL vs MongoDB 10,000 records and 10,000 operations - Throughput")
ax2.set(ylabel='Throughput (ops/sec)', xlabel='Database')
ax2.figure.savefig('/mnt/c/Users/cianw/Documents/Throughput.png', dpi=1200, facecolor='w')
```

```python

fig, (ax3) = plt.subplots(nrows=1, figsize=(15,7))
sns.color_palette("Set2")
sns.set_style("whitegrid")
ax3 = sns.barplot(data=df[df['Variable']=='Insert Avg Latency'], x="Database", y="Result", hue='Workload', palette='Set2', errorbar =None)
ax3.set_title("MySQL vs MongoDB 10,000 records and 10,000 operations - Insert Latency")
ax3.set(ylabel='Average Insert Latency', xlabel='Database')
ax3.figure.savefig('/mnt/c/Users/cianw/Documents/Latency.png', dpi=1200, facecolor='w')
```
